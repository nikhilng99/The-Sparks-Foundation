# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the dataset

```python
df=pd.read_csv('SampleSuperstore.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Dropping the duplicates from the dataset
df = df.drop_duplicates()
```

```python
df.info()
```

# Exploratory Data Analysis

```python
ax=sns.heatmap(df.corr(),annot=True)
ax.set_ylim(sorted(ax.get_xlim(), reverse=True))
```

# Types of Orders placed based on Category

```python
df['Category'].value_counts()
```

```python
sns.countplot(x="Category", data=df).set_title("Most sold product category");
```

```python
sns.catplot('Category',kind='count',hue='Sub-Category',data=df,height=8,aspect=1.5)
plt.title('Sub-Category orders placed',size=20)
```

# Orders by each City

```python
worst_cities=df['City'].value_counts()[::-1][:15]
print(worst_cities)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x=worst_cities,y=worst_cities.index)
plt.title("Bottom 15 Cities (sales wise)",size=20)
plt.xlabel("Number of orders",size=15)
```

```python
df['City'].value_counts()[:10]
```

```python
plt.figure(figsize=(15,7))
abc=df['City'].value_counts()[:10]
sns.barplot(x=abc,y=abc.index)
plt.title("Top Ten Cities (sales wise)",size=20)
plt.xlabel("Number of orders",size=15)
```

# Orders by each State

```python
df['State'].value_counts()[:10]
```

```python
worst_states=df['State'].value_counts()[::-1][:15]
print(worst_states)
```

```python
plt.figure(figsize=(15,7))
abc=df['State'].value_counts()[:10]
sns.barplot(x=abc,y=abc.index)
plt.title("Top 10 states (sales wise)",size=20)
plt.xlabel("Number of orders",size=15)
```

# Most popular Sub-Categories

```python
plt.figure(figsize=(12,7))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# Counts of Sub-Category Region-Wise
plt.figure(figsize=(12,7))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.xticks(rotation=90)
plt.show()
```
